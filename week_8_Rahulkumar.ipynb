# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools
- Add Date and Timne
- Add Count


````python
import re
# 🛠️ TOOL 1: Calculator
def calculator(expression: str) -> str:
    """Evaluate a mathematical expression safely."""
    try:
        # Allow only numbers, operators, parentheses, decimal points, and spaces
        if not re.fullmatch(r"[0-9+\-*/().\s]+", expression):
            return "Invalid mathematical expression"
        result = eval(expression)
        return str(result)
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor
import re
def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        stop_words = {
            "the", "is", "are", "was", "were", "this", "that",
            "with", "from", "into", "about", "have", "has",
            "had", "will", "shall", "would", "could", "should",
            "and", "for", "you", "your", "their", "they",
            "them", "our", "ours", "his", "her", "its",
            "what", "when", "where", "which", "while", "there"
        }
        # Remove punctuation and convert to lowercase
        words = re.findall(r"\b[a-zA-Z]+\b", text.lower())
        # Remove stop words and short words
        keywords = [
            word for word in words
            if len(word) > 4 and word not in stop_words
        ]
        # Remove duplicates while preserving order
        unique_keywords = list(dict.fromkeys(keywords))
        return unique_keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
from datetime import datetime

# Request Counter
request_count = 0

# 🤖 AGENT FUNCTION
def agent(query: str):
    global request_count

    # Increase request count
    request_count += 1

    query_lower = query.lower()

    # Current Date & Time
    current_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    try:
        # 🔢 Calculator Routing
        if "calculate" in query_lower:
            expression = query_lower.replace("calculate", "").strip()

            return {
                "request_id": request_count,
                "type": "calculator",
                "tool": "Calculator",
                "query": query,
                "result": calculator(expression),
                "timestamp": current_time,
                "status": "success"
            }

        # 📝 Keyword Extraction Routing
        elif "keyword" in query_lower or "keywords" in query_lower:
            text = (
                query.replace("Extract keywords from", "")
                     .replace("extract keywords from", "")
                     .strip()
            )

            return {
                "request_id": request_count,
                "type": "keyword_extractor",
                "tool": "Keyword Extractor",
                "query": query,
                "result": extract_keywords(text),
                "timestamp": current_time,
                "status": "success"
            }

        # 💬 General Query
        else:
            return {
                "request_id": request_count,
                "type": "general",
                "tool": "None",
                "query": query,
                "result": f"I received your query: {query}",
                "timestamp": current_time,
                "status": "success"
            }

    except Exception as e:
        return {
            "request_id": request_count,
            "type": "error",
            "tool": "None",
            "query": query,
            "result": str(e),
            "timestamp": current_time,
            "status": "failed"
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````
